1. Data Simulation
Generate synthetic data with realistic ranges for each feature. Add some anomalies that will stand out:


```python
import numpy as np
import pandas as pd

# Set seed for reproducibility
np.random.seed(0)

# Simulating 1000 normal points
data = pd.DataFrame({
    'Threat_Score': np.random.normal(50, 10, 1000),
    'Location_Variance': np.random.normal(5, 2, 1000),
    'Frequency_of_Communication': np.random.normal(3, 1, 1000),
    'Time_of_Day_Activity': np.random.normal(12, 3, 1000), # 24-hour time in military format
    'Activity_Level': np.random.normal(10, 3, 1000)
})

# Add anomalies (e.g., 20 anomalous points)
anomalies = pd.DataFrame({
    'Threat_Score': np.random.normal(85, 5, 20),
    'Location_Variance': np.random.normal(15, 5, 20),
    'Frequency_of_Communication': np.random.normal(1, 0.5, 20),
    'Time_of_Day_Activity': np.random.normal(23, 2, 20),  # Later hours
    'Activity_Level': np.random.normal(30, 5, 20)
})

# Combine normal data with anomalies
data = pd.concat([data, anomalies], ignore_index=True)
```

2. Conduct some Exploratory Dta Analysis in order to understand the generated data.

```python
import matplotlib.pyplot as plt

# List of features to plot
features = ['Threat_Score', 'Location_Variance', 'Frequency_of_Communication', 'Time_of_Day_Activity', 'Activity_Level']

# Set up a figure with subplots for each feature
plt.figure(figsize=(15, 10))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 3, i)  # 2 rows, 3 columns layout
    plt.hist(data[feature], bins=30, edgecolor='k', alpha=0.7)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

# Adjust layout to prevent overlap and show the plot
plt.tight_layout()
plt.show()
```
